### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.


```python
# Exercise 10
import matplotlib.pyplot as plt

# Define x values from 0 to 10
x = list(range(11))

# Define the three functions
y1 = [i for i in x]        # y = x (linear)
y2 = [i**2 for i in x]     # y = x^2 (quadratic)
y3 = [i**3 for i in x]     # y = x^3 (cubic)

# Create a single figure
plt.figure(figsize=(12, 8))

# Plot each function with different colors and line styles
plt.plot(x, y1, color='red', linestyle='-', marker='o', linewidth=2, label='y = x')
plt.plot(x, y2, color='blue', linestyle='--', marker='s', linewidth=2, label='y = x')
plt.plot(x, y3, color='green', linestyle='-.', marker='^', linewidth=2, label='y = x')

# Add labels and title
plt.title('Comparison of Linear, Quadratic, and Cubic Functions', fontsize=16)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values', fontsize=12)

# Add legend to distinguish the three lines
plt.legend(fontsize=12)

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Show the plot
plt.show()
```

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.


```python
# Exercise 16
import matplotlib.pyplot as plt
import numpy as np

# Define the data
categories = ['X', 'Y', 'Z']
group_a = [3, 5, 7]
group_b = [2, 6, 4]

# Set up the bar positions
x = np.arange(len(categories))
width = 0.6  # Width of the bars

# Create the stacked bar chart
plt.figure(figsize=(10, 6))

# Create the first layer (Group A)
bars1 = plt.bar(x, group_a, width, label='Group A', color='#3498db', alpha=0.8)

# Create the second layer (Group B) stacked on top of Group A
bars2 = plt.bar(x, group_b, width, bottom=group_a, label='Group B', color='#e74c3c', alpha=0.8)

# Add labels and title
plt.title('Stacked Bar Chart: Group A and Group B', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)

plt.xticks(x, categories)

# Add legend
plt.legend(fontsize=12)

# Add value labels on each segment
for i, (a, b) in enumerate(zip(group_a, group_b)):
    # Label for Group A (bottom segment)
    plt.text(i, a/2, str(a), ha='center', va='center', fontweight='bold', color='white')
    # Label for Group B (top segment)
    plt.text(i, a + b/2, str(b), ha='center', va='center', fontweight='bold', color='white')
    # Total label above each bar
    plt.text(i, a + b + 0.2, str(a + b), ha='center', va='bottom', fontweight='bold')

# Add grid for better readability
plt.grid(axis='y', alpha=0.3)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

```

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.


```python
# Exercise 17
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 5 for better visualization
x = np.linspace(0, 5, 50)

# Define the four functions
y1 = x          # Linear
y2 = x**2       # Quadratic
y3 = x**3       # Cubic
y4 = x**4       # Quartic

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Define colors for each plot
colors = ['red', 'blue', 'green', 'purple']

# Plot y1 = x in the top-left subplot (0,0)
axes[0, 0].plot(x, y1, color=colors[0], linewidth=2, marker='o', markersize=3)
axes[0, 0].set_title('y = x (Linear)', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
axes[0, 0].grid(True, alpha=0.3)

# Plot y2 = x^2 in the top-right subplot (0,1)
axes[0, 1].plot(x, y2, color=colors[1], linewidth=2, marker='s', markersize=3)
axes[0, 1].set_title('y = x (Quadratic)', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')
axes[0, 1].grid(True, alpha=0.3)

# Plot y3 = x^3 in the bottom-left subplot (1,0)
axes[1, 0].plot(x, y3, color=colors[2], linewidth=2, marker='^', markersize=3)
axes[1, 0].set_title('y = x (Cubic)', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].grid(True, alpha=0.3)

# Plot y4 = x^4 in the bottom-right subplot (1,1)
axes[1, 1].plot(x, y4, color=colors[3], linewidth=2, marker='d', markersize=3)
axes[1, 1].set_title('y = x (Quartic)', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')
axes[1, 1].grid(True, alpha=0.3)

# Add a main title for the entire figure
fig.suptitle('Comparison of Polynomial Functions', fontsize=16, fontweight='bold', y=0.98)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Exercise 19:
Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.


```python
# Exercise 19
import matplotlib.pyplot as plt
import numpy as np

# Define the data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
errors = [1, 1, 1, 1]  # Error of 1 for each value

# Create the bar plot with error bars
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, yerr=errors, capsize=5, 
               color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12'], 
               alpha=0.7, edgecolor='black', linewidth=1)

# Customize error bars
plt.errorbar(categories, values, yerr=errors, fmt='none', 
             ecolor='black', elinewidth=2, capsize=5, capthick=2)

# Add labels and title
plt.title('Bar Plot with Error Bars', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Add value labels on top of each bar
for i, (bar, value, error) in enumerate(zip(bars, values, errors)):
    plt.text(bar.get_x() + bar.get_width()/2, value + error + 0.2, 
             str(value), ha='center', va='bottom', fontweight='bold', fontsize=11)

# Add grid for better readability
plt.grid(axis='y', alpha=0.3)

# Set y-axis limits to accommodate error bars
plt.ylim(0, max(values) + max(errors) + 1)

# Show the plot
plt.tight_layout()
plt.show()
```

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
# Exercise 26
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calculate the z values for the function
Z = X**2 + Y**2

# Create a 3D figure and axes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Create the surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')  # You can choose a different colormap

# Add title and labels
ax.set_title('3D plot of z = x^2 + y^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Show the plot
plt.show()
```

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.


```python
# Exercise 27
import matplotlib.pyplot as plt
import numpy as np

# Define variables and generate random data
variables = ['80', '150', '25', '10', '190']
num_variables = len(variables)
data = np.random.rand(num_variables)  # Random values between 0 and 1

# Calculate angles for each variable
angles = np.linspace(0, 2 * np.pi, num_variables, endpoint=False)

# Close the plot by repeating the first data point and angle
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))

# Set up the polar plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(projection='polar'))

# Plot the data and fill the area
ax.plot(angles, data, 'o-', linewidth=2)
ax.fill(angles, data, alpha=0.25)

# Configure the plot
# Add the first label to the variables list to match the number of angles
ax.set_thetagrids(angles * 180/np.pi, variables + [variables[0]])
ax.set_ylim(0, 1)  # Set the limit for the radial axis
ax.set_title('Radar Chart', va='bottom')

# Show the plot
plt.show()
```

### Exercise 34:
Create a plot with twin y-axes where one axis represents `y = x` and the other represents `y = x^2`.


```python
# Exercise 34
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(0, 10, 100)

# Calculate y values for both functions
y1 = x
y2 = x**2

# Create a figure and a primary axes
fig, ax1 = plt.subplots()

# Plot the first function on the primary axes
ax1.plot(x, y1, 'b-', label='y = x')
ax1.set_xlabel('x')
ax1.set_ylabel('y = x', color='blue')
ax1.tick_params(axis='y', labelcolor='blue') # Set y-axis label and tick color

# Create a twin axes that shares the same x-axis
ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis [1]

# Plot the second function on the twin axes
ax2.plot(x, y2, 'r-', label='y = x^2')
ax2.set_ylabel('y = x^2', color='red')
ax2.tick_params(axis='y', labelcolor='red') # Set y-axis label and tick color

# Add a title to the plot
plt.title('Plot with Twin Y-Axes (y=x and y=x^2)')

# Add legends for both lines
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9)) # Place the legend outside the plot

# Show the plot
plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
import matplotlib.pyplot as plt
import numpy as np

# Example data
categories = ['Start', 'Revenue', 'Costs', 'Investment', 'Profit']
values = [100, 40, -30, -10, 0]  # Profit will be calculated

# Calculate cumulative values for plotting
cumulative = [values[0]]
for v in values[1:-1]:
    cumulative.append(cumulative[-1] + v)

# Add final total (Profit)
values[-1] = cumulative[-1]
cumulative.append(values[-1])

# Calculate bar positions and colors
y = np.array([values[0]] + values[1:-1] + [0])
color = ['gray'] + ['green' if v > 0 else 'red' for v in values[1:-1]] + ['blue']

# Start positions for bars
start = [0]
for i in range(1, len(cumulative)):
    start.append(cumulative[i-1])

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
for i in range(len(values)):
    ax.bar(categories[i], y[i], bottom=start[i], color=color[i])

# Annotate values
for i in range(len(values)):
    ax.text(i, start[i] + y[i] + (2 if y[i] >= 0 else -5), f'{start[i] + y[i]:.0f}',
            ha='center', va='bottom' if y[i] >= 0 else 'top')

# Add title and grid
plt.title('Waterfall Chart Example')
plt.ylabel('Value')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# Exercise 49
import numpy as np
import matplotlib.pyplot as plt

# Attributes and values
labels = ['Speed', 'Strength', 'Agility', 'Intelligence', 'Endurance']
values = [80, 90, 70, 85, 75]

# Number of variables
num_vars = len(labels)

# Compute angle for each axis in the plot (in radians)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is made circular by completing the loop and appending the start to the end
values += values[:1]
angles += angles[:1]

# Create subplot with polar projection
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw the outline of the radar chart
ax.plot(angles, values, color='red', linewidth=2)
ax.fill(angles, values, color='red', alpha=0.25)

# Set the labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Set the range for the radar chart
ax.set_ylim(0, 100)

# Add a title
plt.title('Strengths of Five Attributes', size=15, y=1.1)

plt.show()
```
